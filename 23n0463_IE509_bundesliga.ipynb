```python
import pandas as pd
df=pd.read_csv('bundesliga 2022-23(1) (1).csv',header=0)
del df['Unnamed: 0']
df1=df[:-1]
df1
```

- Split the list (line wise)

---



- Match Team short names to their expanded team names

```python

lists=[]
for i in range(1,19):
  x= list(df1.iloc[i,1:])+list(df1.iloc[1:,i])
  lists.append(x)
print(lists)

goal_for=[]
goal_against=[]
result=[]
for lst in lists:
  for i in lst[0:18]:
    if i=='— ':
      continue
    elif i[0]>i[2]:
      result.append('win')
      goal_for.append(int(i[0]))
      goal_against.append(int(i[2]))
    elif i[0]<i[2]:
      result.append('loss')
      goal_for.append(int(i[0]))
      goal_against.append(int(i[2]))
    elif i[0]==i[2]:
      result.append('draw')
      goal_for.append(int(i[0]))
      goal_against.append(int(i[2]))
    else:
      print("Syntax Error")

  for j in lst[18:37]:
    if j=='— ':
      continue
    elif j[0]<j[2]:
      result.append('win')
      goal_against.append(int(j[0]))
      goal_for.append(int(j[2]))
    elif j[0]>j[2]:
      result.append('loss')
      goal_for.append(int(j[2]))
      goal_against.append(int(j[0]))

    elif j[0]==j[2]:
      result.append('draw')
      goal_for.append(int(j[2]))
      goal_against.append(int(j[0]))

AUG=result[0:34]
BSC=result[34:68]
UNB=result[68:102]
BOC=result[102:136]
BRE=result[136:170]
DOR=result[170:204]
FRA=result[204:238]
FRE=result[238:272]
HOF=result[272:306]
KOL=result[306:340]
LEI=result[340:374]
LEV=result[374:408]
MAI=result[408:442]
MON=result[442:476]
MUN=result[476:510]
SCH=result[510:544]
STU=result[544:578]
WOL=result[578:612]



```

- Print the win/loss for all games played by a single team
